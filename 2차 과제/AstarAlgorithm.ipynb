```python
from math import sqrt
from queue import PriorityQueue

start='806524731'
goal='123456780'

n=int(sqrt(len(goal))) # 보드 크기

class Node:
    def __init__(self, state: str, cur_price: int = 0):
        self.state=state
        self.cur_price = cur_price

    def f_mismatch(self):#불일치 수 검사하는 함수
        
        return sum(self.state[i]!=goal[i] for i in range(len(goal)))

    def f_manhattan(self):#f'(n)=g(n)+h'(n)에서 h'(n). 목표까지 예상비용
        dist=0
        for i in range(len(goal)):
            r,c=i//n,i%n
            r1,c1=self.state.index(goal[i])//n,self.state.index(goal[i])%n
            dist+=abs(r1-r)+abs(c1-c)
        return dist
    def g(self):
        return self.cur_price
   
    def f(self):#예상 경로 비용
        return self.g()+self.f_manhattan()

    def __lt__(self,other):
        return self.f()<other.f()

def print_puzzle(state, cnt):
    if state == start:
        print("Initial State:")
    elif state == goal:
        print("Goal state:")
    else:
        print("Step:\t", cnt)
    for i in range(n):
        print(state[n*i:n*(i+1)])
        
def exchange(node:Node, pre_position:int, new_position:int):
    new_state=list(node.state)
    new_state[pre_position], new_state[new_position]=new_state[new_position], new_state[pre_position]
    return Node(''.join(new_state), node.cur_price+1)

def Move(node: Node):
    res=[]
    p=node.state.index('0') # 빈 칸 위치
    child=[]
    if not (p%n==0): # 좌변이 아니면
        child.append(exchange(node, p, p-1))
    if not (p%n==n-1): # 우변이 아니면
        child.append(exchange(node, p, p+1))
    if p>=n: # 상변이 아니면 
        child.append(exchange(node, p, p-3))
    if p<n*n-n: # 하변이 아니면 
        child.append(exchange(node, p, p+3))
    return child

def print_path(child):
    state=[]
    nd=goal
    cnt=0
    while nd != start:
        state.append(nd)
        nd=child[nd]
    state.append(start)
    for each in state[::-1]:
        print_puzzle(each, cnt)
        cnt+=1
        
node=Node(start)
Q=PriorityQueue()
Q.put(node)
V={node.state: None}
while not Q.empty():
    cur_node=Q.get()
    if cur_node.state==goal:
        break
    else:
        for node in Move(cur_node):
            if node.state not in V:
                Q.put(node)
                V[node.state]=cur_node.state
print_path(V)
```
